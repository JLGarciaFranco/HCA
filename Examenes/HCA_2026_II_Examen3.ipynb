## Introducción

En el estudio de crecimiento de poblaciones, limitadas por competencia por alimentos, se tiene un modelo matemático que se expresa en términos de una ecuación diferencial ordinaria como sigue:

$$
\frac{d P(t)}{d t} =  r P ( 1 - P / K) \; \text{ para } \; t > 0 \tag{1}
$$

esta ecuación se completa con la siguiente condición inicial:
$$
P(0) = P_0
$$

donde:

* $P_0$ es la población inicial,
* $r > 0$  es la tasa de crecimiento intrínseco de la población y
* $K$ es la capacidad de carga del entorno.

Este modelo se basa en la idea de que el crecimiento de una población comienza de forma exponencial, pero se ralentiza a medida que la población se acerca a la capacidad de carga del entorno $K$.

Obsérvese que en este caso la ecuación es no lineal pues del lado derecho aparece un término cuadrático de $P$.

Esta ecuación tiene la siguiente solución analítica (conocida como la función logística):
$$
P(t) = \dfrac{K}{1 + \left(\dfrac{K-P_0}{P_0}\right)e^{-rt}} \tag{2}
$$
