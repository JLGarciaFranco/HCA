## **Parcial 1**

1. Sucesión de Fibonacci

La sucesión de Fibonacci se define de la siguiente manera:
 $
F_0 = 0, \qquad F_1 = 1
$

y, para $n \geq 2$, $F_n = F_{n-1} + F_{n-2}$

Es decir, cada término de la sucesión se obtiene sumando los dos términos anteriores. Los primeros términos son:

$
0, 1, 1, 2, 3, 5, 8, 13, 21, \ldots
$

  *1.1* Escribe un programa que genere los primeros `N` números de la sucesión de Fibonacci.

Tu programa debe pedir al usuario un número entero positivo `N`, generar los primeros `N` términos de la sucesión de Fibonacci, guardar los términos en una lista, imprimir la lista resultante.

  *1.2* Escribe un programa que encuentre los primeros 5 números primos que aparecen en la sucesión de Fibonacci.
