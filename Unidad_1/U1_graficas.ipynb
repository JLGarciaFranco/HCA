 # **<font color="ForestGreen">  Matplotlib: qué es y por qué usarlo </font>**
 


Matplotlib es la biblioteca básica de gráficos en Python. Permite dibujar líneas, puntos, barras y figuras con control sobre ejes, etiquetas y formato. Muchas herramientas (pandas `.plot`, seaborn, xarray, cartopy) se apoyan en ella, por lo que aprenderla te da una base común.

Por qué es útil:
- Se integra bien con NumPy y pandas (series temporales, tablas).
- Da control fino para figuras “listas para informe”.
- Funciona para un panel o múltiples subgráficos.
- Exporta a PNG/PDF/SVG de forma reproducible.



Cómo pensarla:
- Figura (lienzo) y ejes (donde dibujas).
- Dos estilos: rápido con `plt.plot(...)` o orientado a objetos con `fig, ax = plt.subplots(); ax.plot(...)` para proyectos más claros.

### Algunas instrucciones de pyplot para enchular la gráfica

`plt.figure`: Crear objeto de figura. Usualmente sirve para especificar el tamaño que va a tener nuestro objeto de figura.

`plt.plot`: Para trazar gráficos de línea.

`plt.xlabel`: Para establecer la etiqueta del eje x.

`plt.ylabel`: Para establecer la etiqueta del eje y.

`plt.title`: Para establecer el título de la gráfica.

`plt.legend`: Para agregar leyendas a la gráfica.

`plt.grid`: Para activar o desactivar las líneas de la cuadrícula.

`plt.xlim`: Para establecer los límites del eje x.

`plt.ylim`: Para establecer los límites del eje y.

`plt.xticks`: Para establecer las marcas del eje x.

`plt.yticks`: Para establecer las marcas del eje y.

````python
# Ejemplo de funciones trigonometrics
# Generar un arreglo en radianes utilizando numpy
angles = np.linspace(-np.pi, np.pi, 100)

# Calcula el valor de los senos y cosenos
sin_values = np.sin(angles)
cos_values = np.cos(angles)

# Grafique los resultados
# Clase documentar cada linea

plt.plot(angles, sin_values, label='sen(x)')
plt.plot(angles, cos_values, label='cos(x)')
# title define el titulo
plt.title('Funciones trigonométricas')
# xlabel define la etiqueta del eje "x"
plt.xlabel('Ángulo (radianes)')
# ylabel define la etiqueta del eje "y"
plt.ylabel('Función')
# la instruccion legend agrega una leyenda
plt.legend()
# grid con lineas punteadas 
plt.grid(linestyle='--')
````

````python
# Calcular el valor de la tangente para los angulos definidos anteriormente
tan_values = np.tan(angles)

# Graficar la tangente, primero modificando el tamano de la figura figsize
plt.figure(figsize=(8, 6))

# modificamos tambien las lineas
plt.plot(angles, tan_values, label='tan(x)',linestyle='--',linewidth=2)
plt.plot(angles, sin_values/cos_values, label='tan2(x)',linestyle='-.',linewidth=2)
plt.xlim([-3,3])
plt.title('Funciones trigonométricas')
plt.xlabel('Ángulo (radianes)')
plt.ylabel('Función')
plt.legend()
plt.grid(True)
````

## Parámetros en gráficos de línea

- **Linewidth** (`linewidth`): Este parámetro controla el grosor de la línea en el gráfico. Puedes especificar el grosor de la línea utilizando un valor numérico. Por ejemplo, linewidth=2 establecerá el grosor de la línea en 2 puntos. 

- **Color** (`color` o `c`): Este parámetro controla el color de la línea en el gráfico. Puedes especificar el color utilizando nombres de colores estándar de matplotlib (como 'blue', 'red', 'green', etc.), códigos hexadecimales ('#FF5733'), o abreviaturas de color ('b' para azul, 'r' para rojo, 'g' para verde, etc.) Para ver todos los colores, puedes consultar: https://matplotlib.org/stable/gallery/color/named_colors.html

- **Linestyle** (`linestyle`): Este parámetro controla el estilo de la línea en el gráfico. Puedes especificar diferentes estilos de línea, como sólida ('-'), discontinua ('--'), punteada (':'), o guiones y puntos ('-.'). 

Además, podemos modificar más detalles de la gráfica como el tamaño de los ejes, de las etiquetas y de los títulos. 

Dentro de `plt.xlabel`, `plt.ylabel` y `plt.title` podemos modificar el tamaño de la letra con la palabra (key) fontsize y luego un número que es el número de puntos de la tipografía.

Además, podemos modificar la leyenda con diferentes kwargs como shadow. Para ver todos los cambios posibles, ver: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html.

````python
x=np.arange(0,3,0.01)
# Calcular la funcion exp(x)
expx=np.exp(x)
# Calcular la funcion log(x)
logx=np.log(x)

# Graficar la funcion exp

plt.figure(figsize=(8, 6))
plt.plot(x,expx, label='exp(x)',linestyle='--')
plt.plot(x,logx, label='log(x)',linestyle='-.')
plt.title('Otras funciones título grande',fontsize=16)
plt.xlabel('Ángulo (radianes)',fontsize=14)
plt.ylabel('Función',fontsize=14)
plt.minorticks_on()
plt.legend(fancybox=True,shadow=True)
plt.grid(True)
plt.show()
````

Ahora podemos hacer una función para poder tenerla lista para todas nuestras gráficas de línea. Esta función la llamaremos `grafico_linea_ts`. En esta función tendremos argumentos obligatorios (*args) y opcionales (*kwargs), acorde al criterio del profe pero que usted puede decidir cambiar. 

````python
def grafico_linea_ts(xmin,xmax,xlab,ylab,title,ymin=None,ymax=None,legend=False):
    # especificar limites minimos y maximos del grafico
    plt.xlim([xmin,xmax])
    plt.ylim([ymin,ymax])
    # cuadrilla interna del grafico
    plt.grid(linestyle='--',alpha=0.66)
    plt.minorticks_on()
    # hacer las etiquetas de ambos ejes mas grandes
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    # poner las labels de los ejes y el titulo
    plt.xlabel(xlab,fontsize=14)
    plt.ylabel(ylab,fontsize=14)
    plt.title(title,fontsize=15)
    # checar si queremos leyenda, si sí, la ponemos y fancy
    if legend:
        plt.legend(fontsize=12.5,fancybox=True,shadow=True)
````

````python
x=np.arange(0,3,0.01)
# Calcular la funcion exp(x)
expx=np.exp(x)
# Calcular la funcion log(x)
logx=np.log(x)

# Graficar la funcion exp

plt.figure(figsize=(8, 6))
plt.plot(x,expx, label='exp(x)',linestyle='--',c='navy',linewidth=2)
plt.plot(x,logx, label='log(x)',linestyle='-.',c='forestgreen',linewidth=2)
grafico_linea_ts(0,3,'x','f(x)','Grafico con nuestra funcion',legend=True)
````

Para aplicar un estilo **globalmente** (afecta todos los gráficos del notebook):
```python

```

Para aplicar un estilo **solo temporalmente** a un bloque de código, sin afectar el resto:
```python

    plt.plot(x, y)
    plt.show()
```

Para regresar al estilo por defecto, puedes usar `'default'`, pero **¡con cuidado!**: esto restablece *todos* los parámetros globales de matplotlib (`rcParams`), incluyendo tamaños de figura, fuentes y colores que hayas configurado manualmente antes:

### **<font color="ForestGreen"> Gráficos de contorno </font>**

Los gráficos de contorno son una herramienta visual eficaz para representar datos bidimensionales en forma de mapas de calor. En Matplotlib, hay dos enfoques comunes para crear gráficos de contorno: utilizando `imshow` y `plt.contourf`. La función imshow permite representar datos como una imagen en escala de grises o en colores, donde los niveles de intensidad indican los valores numéricos.

Para poder graficar algo en 2D, necesitamos un arreglo de 2D. Para generar un arreglo 2D podemos utilizar la función de numpy `meshgrid`. 

La función `np.meshgrid` en NumPy se utiliza para crear matrices de coordenadas a partir de matrices unidimensionales, lo que es particularmente útil para generar mallas de puntos en 2D para su posterior uso en gráficos tridimensionales, interpolación de datos, y cálculos vectoriales.

Cuando se aplica `np.meshgrid` a dos o más arrays unidimensionales, devuelve matrices multidimensionales donde cada dimensión representa un conjunto de coordenadas. Por ejemplo, si tienes dos arrays unidimensionales `x` y `y`, `np.meshgrid(x, y)` generará dos matrices bidimensionales, una correspondiente a las coordenadas $x$ y otra a las coordenadas $y$.

Las salidas de `np.meshgrid` son útiles para operaciones que involucran manipulación de coordenadas, como trazado de gráficos en 3D, evaluación de funciones en una cuadrícula de puntos, interpolación de datos, cálculos de derivadas parciales y más.

En resumen, `np.meshgrid` es una función fundamental para la creación de mallas de puntos en 2D a partir de arrays unidimensionales, lo que facilita la representación visual de datos y cálculos vectoriales en el espacio bidimensional.

Para ver un ejemplo y calcularlo:

````python
# Crear datos bidimensionales
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)

# Meshgrid genera arreglos combinando los valores de x y y.
X, Y = np.meshgrid(x, y)

# sqrt es la raiz cuadrada en numpy

Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear gráfico de contorno con imshow
plt.imshow(Z, extent=[-2, 2, -2, 2], origin='lower', cmap='viridis')
plt.colorbar(label='Valor')
plt.title('Gráfico de Contorno con imshow')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
````

En este ejemplo, imshow representa la función sinusoidal bidimensional $Z(X,Y)$ como un gráfico de contorno, donde los colores indican los valores de la función.

La función `plt.contourf`, por otro lado, genera gráficos de contorno rellenados, creando áreas de diferentes colores para representar niveles de datos. Aquí hay un ejemplo:

````python
# Crear gráfico de contorno relleno con plt.contourf
contorno_relleno = plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(contorno_relleno, label='Valor')
plt.title('Gráfico de Contorno Relleno con plt.contourf')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
````

Al emplear `plt.contourf`, se puede especificar un conjunto de datos bidimensional, generalmente representado como una matriz, y la función generará un mapa de contorno coloreado donde los colores indican diferentes niveles de valores en la matriz. Además, es posible personalizar la cantidad de contornos, los colores utilizados y otros aspectos visuales para adaptar el gráfico a las necesidades específicas del análisis.

La palabra `levels` por ejemplo nos permite indicar los niveles específicos del gráfico.

También podemos cambiar los colores a través de la palabra *colormap*, por ejemplo usamos el mapa **viridis** pero podríamos cambiarlo, aquí están todas las opciones 

````python
# Crear gráfico de contorno relleno con plt.contourf
contorno_relleno = plt.contourf(X, Y, Z, cmap='viridis',levels=np.arange(0,1,0.01))
plt.colorbar(contorno_relleno, label='Valor')
plt.title('Gráfico de Contorno Relleno con plt.contourf')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
````
