## Nested if (if anidados)

Un nested `if` es una estructura condicional dentro de otra. Es decir, colocas un `if` (o `elif`, o `else`) dentro del bloque de otro `if`. Esto te permite evaluar condiciones dependientes o jer谩rquicas.

### *Cuidado*
- No anides demasiados niveles, o el c贸digo se vuelve dif铆cil de mantener.
- Considera usar `elif` o funciones si la l贸gica se vuelve muy compleja.


````python
temperatura = 2
humedad = 80
viento = 51

if temperatura < 5:
    print("Temperatura baja")
    if humedad >= 80:
        print("Riesgo de formaci贸n de niebla")
        if viento > 50:
            print("Riesgo de supersaturacion")
    else:
        print("Clima fr铆o pero seco")
else:
    print("Temperatura moderada")
````

## Operador ternario

Se trata de una cl谩usula `if`, `else` que se define en una sola l铆nea y puede ser usado por ejemplo, dentro de un `print()`

```
# [c贸digo si se cumple] if [condici贸n] else [c贸digo si no se cumple]
```

````python
x = 5
print("Es 5" if x == 5 else "No es 5")
````
